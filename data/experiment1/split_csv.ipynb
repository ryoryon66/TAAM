```python
import pandas as pd
```

```python
df_all = pd.read_csv("time_data.csv")
df_all.shape
```

```python
df_all.describe()
df_all.drop(len(df_all) - 1,axis=0)

# eliminate inappropriate data
for col in df_all.columns:
    df_all = df_all[df_all[col] > -0.001]
    
```

```python
df_all.describe()
```

```python
def split_df(df_all,k=2):
    
    dfs = []
    
    for i in range(k):
        df = df_all[i * len(df_all) // k:(i+1) * len(df_all) //k]
        dfs.append(df)
    
    return dfs

df1,df2,df3 = tuple(split_df(df_all,3))
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

```python
def save_csvs(df_all,k=2):
    dfs = split_df(df_all,k)
    
    for i,df in enumerate(dfs):
        i += 1
        df.to_csv(f"mainPC_time_data{i}.csv",index= False)
        print(df.shape)
    
    return
```

```python
save_csvs(df_all,3)
```
